# In this notebook we will add the predictions to the dataset and make our final selection

```python
%matplotlib inline
import numpy as np
import pandas as pd
import joblib
df = pd.read_csv('datasets/score/score_cleaned.csv')
```

#### Damage Classifier

```python
gbm_classifier = joblib.load('pickles/GBM_ClassifierFinal.pkl')
```

```python
pred_dmg_class = gbm_classifier.predict(df)
```

```python
print(pred_dmg_class)
```

#### Profit Amount

```python
profit_best_model = joblib.load('pickles/Profit_best_model.pkl')
```

```python
pred_profit = profit_best_model.predict(df)
```

```python
print(pred_profit)
```

#### Damage Amount

```python
damage_best_model = joblib.load('pickles/Damage_amount_best_model.pkl')
```

```python
pred_dmg = damage_best_model.predict(df)
```

```python
print(pred_dmg)
```

```python
print(np.expm1(pred_profit))
```

```python
df['Dmg_class'] = pred_dmg_class
```

```python
df['Dmg_amount'] = np.expm1(pred_dmg)
```

```python
df['Profit_amount'] = np.expm1(pred_profit)
```

```python
df.to_csv('datasets/score/scoreWithPredictions.csv', index=False)
```

```python
df = pd.read_csv('datasets/score/scoreWithPredictions.csv')
```

## Selection
To select the final list we take the netprofit by substracting the dmg amount from the profit amount if the guest is predicted to do damage.

```python
df['NetProfit'] = df['Profit_amount'] - df['Dmg_amount']
```

```python
df.hist(column='Dmg_class')
```

```python
df.head()
```

```python
df['NetProfitWithClass'] = np.where(df['Dmg_class']==0,df['Profit_amount'],df['Profit_amount']-df['Dmg_amount'])
```

```python
df.head()
```

Added ID column to have some easier identification

```python
df['ID'] = range(1, len(df) + 1)
```

```python
df = df.rename(columns = {'Dmg_class':'Predicted_Damage_Status','Dmg_amount':'Predicted_Damage_Cost','Profit_amount':'Predicted_Revenue','NetProfitWithClass':'Overall_Predicted_Revenue'})
```

```python
df.drop('NetProfit', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.to_csv('datasets/score/FinalApplicants.csv', index=False)
```

Sort and select the top 200 applicants

```python
df_sorted = df.sort_values('Overall_Predicted_Revenue',ascending=False)
```
